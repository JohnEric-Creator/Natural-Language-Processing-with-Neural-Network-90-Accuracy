**Common data cleaning steps on all text:**
* Make text all lower case
* Remove punctuation
* Remove numerical values
* Remove common non-sensical text (/n)
* Tokenize text
* Remove stop words

**More data cleaning steps after tokenization:**
* Stemming / lemmatization
* Parts of speech tagging

###### First round of text cleaning techniques
Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.

```python
def clean_text_round1(data):
    data = data.lower()
    data = re.sub('\[.*?\]', '', data)
    data = re.sub('[%s]' % re.escape(string.punctuation), '', data)
    data = re.sub('\w*\d\w*', '', data)
    return data


round1 = lambda x: clean_text_round1(x)
```

```python
data_clean = pd.DataFrame(data.Review.apply(round1))
data_clean
```

###### Second round of cleaning
Eliminate some additional punctuation and non-sensical text that was missed the first time around

```python
def clean_text_round2(data):
    data = re.sub('[‘’“”…]', '', data)
    data = re.sub('\n', '', data)
    return data

round2 = lambda x: clean_text_round2(x)
```

###### Observing the restructured text

```python
data_clean = pd.DataFrame(data_clean.Review.apply(round2))
data_clean
```

Removing any URL’s,create clean_review column, and clean text of unnecessary punctuation

```python
def clean_url(data_clean):
    return re.sub('r.http\S+', '', data_clean)


data['Prepared_Review'] = data_clean['Review'].apply(clean_url)


def clean(txt):
    txt = txt.str.replace('(<br/>)', '')
    txt = txt.str.replace('(<a).*(>).*(</a>)', '')
    txt = txt.str.replace('(&amp)', '')
    txt = txt.str.replace('(&gt)', '')
    txt = txt.str.replace('(&lt)', '')
    txt = txt.str.replace('(\xa0)', '')
    return txt


# Refresh clean_review column
data['Prepared_Review'] = clean(data['Prepared_Review'])
```

###### Proposed Word Embedding Length

In research from Pennington (2014), The GloVe model is trained on the non-zero records of a global word-word co- circumstance matrix, which formulates how regularly words co-occur with each another in each corpus. Filling this matrix requires a solitary pass through the complete corpus to amass the statistics. For big corpora, this pass can be computationally costly, but it is a one-time up-front expenditure. Consequent training iterations are much quicker since the number of non-zero matrix entries are characteristically much smaller than the totality of the number of words in the corpus. 

The tools afforded in this package automate the collection and provision of co-occurrence statistics for contribution into the model. The core training code is removed from these preprocessing steps and can be executed autonomously. 

The training objective of GloVe is to learn word vectors such that their dot product equals the logarithm of the words' probability of co-occurrence. Because the logarithm of a ratio equals the difference of logarithms, this objective associate (the logarithm of) ratios of co-occurrence probabilities with vector differences in the word vector space. Because these ratios can encode some form of meaning, this information gets encoded as vector differences as well. For this reason, the resulting word vectors perform very well on word analogy tasks, such as those examined in the word2vec package.

To capture in a quantitative way, the nuance necessary to distinguish word from word, it is necessary for a model to associate more than a single number to the word pair. A natural and simple candidate for an enlarged set of discriminative numbers is the vector difference between the two-word vectors. Word2vec is designed in order that such vector differences capture as much as possible the meaning specified by the juxtaposition of two words.  

Word2vec is what we will be utilizing for this analysis

## B.2.   Tokenization and Tokenization Goals

According to research from (Gupta, 2021) tokenization is breaking a portion of text into smaller parts. Whether, breaking a paragraph into sentences or sentence into words or a word into characters.  We understand as we read each word, comprehending its meaning, then read the subsequent word until we reach a full stop. 

Accordingly, the model also needs all words distinctly and sequentially that comprise the sentence. So, if we are given a paragraph, we need to get all sentences. From all these sentences, we need words & then only we can move on for any sort of comprehension and prediction.

Tokens are the substance of Natural Language, corresponding to research from (Pai, 2021), the most usual way of processing raw text transpires at the token level.  For example, Transformer based models which are considered State of The Art (SOTA) Deep Learning architectures, NLP processes the text at the token level.  Similarly, the most predominant deep learning architectures for NLP like RNN, GRU, and LSTM also process raw text at the token level.

Tokenization is the fundamental step while modeling text data. Tokenization is executed on the corpus to obtain tokens. The subsequent tokens are then sourced to formulate a vocabulary. Vocabulary denotes to the set of unique tokens in the corpus. Vocabulary is constructed by considering each unique token in the corpus or by considering the top frequently appearing words.  Establishing Vocabulary is the primary goal of Tokenization, the simplest way to increase the performance of the NLP model is to generate a vocabulary out of top K frequently occurring words.

## C.3.   Rationalization for Hyperparameter Elections

Concerning the model’s ‘Relu’ activation function
* According to research from Brownlee (2021) the ‘Relu” (rectified linear activation function ) is possibly the most popular function exercised for hidden layers.  It is commonplace since it is mutually uncomplicated to implement and efficient at prevailing over the drawbacks of other formerly popular activation functions, like Sigmoid and Tanh. Specifically, it is less vulnerable to vanishing gradients that avert deep models from being trained, though it can suffer from other problems like saturated or “dead” units.

Concerning the model’s binary cross entropy loss function
* Cross-entropy loss refers to the contrast between two random variables; it measures them to extract the difference in the information they contain, showcasing the results, according to research (Saxena, 2021).  This type of loss function is used to calculate how accurate the machine learning or deep learning model is by defining the difference between the estimated probability with the desired outcome.  Essentially, this type of loss function measures the model’s performance by transforming its variables into real numbers, thus, evaluating the “loss” that’s associated with them. The higher the difference between the two, the higher the loss.  Cross-entropy loss is utilized in many classification tasks it is the most popular loss function for such cases.  While the outputs in regression tasks are numbers, the outputs for classification are binary categories, like cats and dogs or true and false.  Which is just what is required for this analysis. 

Concerning the model’s number of nodes per layer
* In research from Brownlee (2019) nodes are also known as Perceptron.  And when it comes to Perceptron, I experimented with different models, running it with as few as one perceptron, to as many as one hundred, and many in-between.  I was able to achieve a range high of 90.29% and a low of 52%.  I kept the high with 5 nodes.

Concerning the model’s binary cross entropy loss function
* In research from Godoy (2019) ff we fit a model to perform a binary classification, it will predict a probability of each one of the points. Given what we know about the points, how can we evaluate how good (or bad) are the predicted probabilities? This is the whole purpose of the loss function! It should return high values for bad predictions and low values for good predictions.  And this is precisely what the binary cross entropy loss function executes.  In addition to this endorsement for Godoy, he blatantly states If you are training a binary classifier, chances are you are using binary cross-entropy / log loss as your loss function, as it is currently the favorite for evaluating probabilistic loss.  


Concerning the models ‘adam’ optimizer
* According to new research from (Bester, 2021), Adam, derived from Adaptive Moment Estimation, is an optimization algorithm.  The Adam optimizer has seen widespread adoption among the deep learning community. Currently the Adam optimizer is the preferred optimizer for use with deep learning models.

Concerning the models Stopping Criteria ‘Early Stopping’ 
* According to new research (Goswami, 2021), overfitting is a very serious problem for all machine learning and deep learning problems. You can get to understand this is happening when your model is doing well on training data, but not able to replicate that performance on test data.  The general set of strategies against this curse of overfitting is called regularization and early stopping is one such technique.  The idea is very simple. The model tries to chase the loss function crazily on the training data, by tuning the parameters. Now, we keep another set of data as the validation set and as we go on training, we keep a record of the loss function on the validation data, and when we see that there is no improvement on the validation set, we stop, rather than going all the epochs. This strategy of stopping early based on the validation set performance is called Early Stopping.  Early Stopping, hence, does not only protect against overfitting but needs considerably a smaller number of Epoch to train.


Concerning the models Evaluation Metric ‘Accuracy’
* In research from Nicholson (2020) accuracy is overall, how often is the model correct.  As a heuristic, or rule of thumb, accuracy can tell us immediately whether a model is being trained correctly and how it may perform generally. However, it does not give detailed information regarding its application to the problem.


```python
num_epochs = 20
history = model.fit(padded,
 training_labels_final,
epochs = num_epochs,
                     callbacks = EarlyStopping(monitor = 'val_loss'), 
 validation_data = (testing_padded, test_labels_final))
```
